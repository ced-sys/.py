```python
import matplotlib.pyplot as plt

def draw_fishbone_diagram():
    # Create figure and axis with larger size
    fig, ax = plt.subplots(figsize=(15, 8))
    ax.set_xlim(-1.2, 1.2)
    ax.set_ylim(-1, 1)
    ax.axis('off')

    # Draw main spine (central line) - made slightly shorter
    ax.plot([0, 0], [-0.7, 0.7], color='black', linewidth=3)

    # Define categories and their associated causes
    categories = {
        "URL Obfuscation": ["Encoded URLs", "Special characters", "Hidden redirects"],
        "Domain Tricks": ["Short TLD length", "Typosquatting", "Look-alike domains"],
        "Similarity Index": ["Minor variations", "Mimic real sites"],
        "User Behavior": ["Lack of awareness", "Clicking blindly"],
        "Technical Factors": ["High obfuscation ratio", "Redirect chains", "URL encoding tricks"],
    }

    # Adjusted y-positions for better spacing
    y_positions = [-0.6, -0.3, 0, 0.3, 0.6]

    for i, (category, causes) in enumerate(categories.items()):
        y_pos = y_positions[i]

        # Draw branch lines from spine to categories - made longer
        ax.plot([-0.8, 0], [y_pos, y_pos], color='black', linewidth=2)

        # Add category labels - improved positioning and size
        ax.text(-0.85, y_pos, category, fontsize=14, fontweight='bold',
                verticalalignment='center', horizontalalignment='right')

        # Add cause labels with improved spacing and alignment
        for j, cause in enumerate(causes):
            # Alternate causes above and below the branch line for better readability
            offset = 0.06 if j % 2 == 0 else -0.06
            ax.text(-0.75, y_pos + offset, f"• {cause}",
                   fontsize=12, verticalalignment='center',
                   horizontalalignment='left')

    # Add main problem label - improved positioning and size
    ax.text(0.1, 0, "Phishing\nAttacks", fontsize=16, fontweight='bold',
            verticalalignment='center', horizontalalignment='left')

    # Add a light grid background for better visual organization
    plt.grid(True, alpha=0.15, linestyle='--')

    # Adjust layout to prevent text cutoff
    plt.tight_layout()
    plt.show()

# Generate the diagram
draw_fishbone_diagram()
```

```python
import matplotlib.pyplot as plt

def draw_fishbone_diagram():
    # Create figure and axis with refined dimensions
    fig, ax = plt.subplots(figsize=(16, 8))
    ax.set_xlim(-1.5, 1)
    ax.set_ylim(-1, 1)
    ax.axis('off')

    # Draw main spine
    ax.plot([0, 0], [-0.7, 0.7], color='black', linewidth=2.5)

    # Define categories and their associated causes
    categories = {
        "Technical Factors": ["High obfuscation ratio", "Redirect chains", "URL encoding tricks"],
        "User Behavior": ["Lack of awareness", "Clicking blindly"],
        "Similarity Index": ["Minor variations", "Mimic real sites"],
        "Domain Tricks": ["Short TLD length", "Typosquatting", "Look-alike domains"],
        "URL Obfuscation": ["Encoded URLs", "Special characters", "Hidden redirects"],
    }

    # Evenly space the categories
    y_positions = [-0.6, -0.3, 0, 0.3, 0.6]

    for i, (category, causes) in enumerate(categories.items()):
        y_pos = y_positions[i]

        # Category label on the left
        ax.text(-1.3, y_pos, category,
                fontsize=12, fontweight='bold',
                verticalalignment='center',
                horizontalalignment='right',
                bbox=dict(facecolor='white', edgecolor='none', pad=5))

        # Draw branch line
        ax.plot([-1.1, 0], [y_pos, y_pos], color='black', linewidth=1.5)

        # Add causes with proper spacing
        for j, cause in enumerate(causes):
            x_pos = -1.0 + (j * 0.3)  # Space causes horizontally
            ax.text(x_pos, y_pos + 0.05, '•',
                   fontsize=12, verticalalignment='center',
                   horizontalalignment='center')
            ax.text(x_pos + 0.02, y_pos - 0.02, cause,
                   fontsize=10, verticalalignment='top',
                   horizontalalignment='left')

    # Main problem label on the right
    ax.text(0.1, 0, "Phishing\nAttacks",
            fontsize=14, fontweight='bold',
            verticalalignment='center',
            bbox=dict(facecolor='white', edgecolor='black', pad=10))

    # Remove grid and add subtle background color
    ax.set_facecolor('#f8f9fa')

    # Adjust layout
    plt.tight_layout(pad=2.0)
    plt.show()

# Generate the diagram
draw_fishbone_diagram()
```
