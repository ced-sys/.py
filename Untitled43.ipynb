```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Constants (arbitrary units for visualization)
earth_orbit_radius = 1.0
mars_orbit_radius = 1.52
earth_period = 1.0         # Earth year
mars_period = 1.88         # Mars year

# Time parameters
num_frames = 360
times = np.linspace(0, 2 * np.pi, num_frames)

# Set up plot
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect('equal')
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_title("Apparent Retrograde Motion of Mars")

# Objects
sun, = ax.plot(0, 0, 'yo', markersize=12, label='Sun')
earth_dot, = ax.plot([], [], 'bo', label='Earth')
mars_dot, = ax.plot([], [], 'ro', label='Mars')
mars_trail, = ax.plot([], [], 'r--', linewidth=1, alpha=0.5, label='Apparent Mars Path')

# Apparent Mars path from Earth
apparent_path_x = []
apparent_path_y = []

def update(frame):
    t = times[frame]

    # Planet positions
    earth_angle = 2 * np.pi * t / earth_period
    mars_angle = 2 * np.pi * t / mars_period

    x_earth = earth_orbit_radius * np.cos(earth_angle)
    y_earth = earth_orbit_radius * np.sin(earth_angle)

    x_mars = mars_orbit_radius * np.cos(mars_angle)
    y_mars = mars_orbit_radius * np.sin(mars_angle)

    # Apparent position of Mars from Earth (relative vector)
    apparent_x = x_mars - x_earth
    apparent_y = y_mars - y_earth

    earth_dot.set_data(x_earth, y_earth)
    mars_dot.set_data(x_mars, y_mars)

    # Collect trail points for apparent Mars path
    apparent_path_x.append(apparent_x)
    apparent_path_y.append(apparent_y)
    mars_trail.set_data(apparent_path_x, apparent_path_y)

    return earth_dot, mars_dot, mars_trail

ani = FuncAnimation(fig, update, frames=num_frames, interval=50, blit=True)
plt.legend()
plt.show()
```
